#Nice Curves

```julia
using Gadfly
```

```julia
function plotc(x, y, title)
    plot(x = x, y = y, Geom.path,
         Coord.cartesian(fixed=true),
         Guide.title(title),
         Theme(default_color=color("tomato")))
end
```

## Fermat's Spiral
In polar coordinates:
\begin{equation}
r = \pm \sqrt{\theta}
\end{equation}

```julia
t = [-6:0.01:6]
x = sign(t).*sqrt(abs(t)).*cos(abs(t))
y = sign(t).*sqrt(abs(t)).*sin(abs(t))

plotc(x, y, "Fermat's Spiral")
```

#Butterfly Curve
\begin{align}
x &= \sin(t) \left[ e^{\cos(t)} - 2\cos(4t) - \sin^5 \left( \frac{t}{12} \right) \right]\\
y &= \cos(t) \left[ e^{\cos(t)} - 2\cos(4t) - \sin^5 \left( \frac{t}{12} \right) \right]
\end{align}

```julia
t = [-8:0.01:8]
a = exp(cos(t)) - 2.0.*cos(4.0.*t) - (sin(t./12.0)).^5.0
x = a.*sin(t)
y = a.*cos(t)

plotc(x, y, "Butterfly Curve")
```

##A Heart Curve


```julia
t =[0.0:0.01:2]
x = 16.0.*(sin(t).^3.0)
y = 13.0.*cos(t) - 5.0.*cos(2.0.*t) - 2.0.*cos(3.0.*t) - cos(4.0.*t)

plotc(x, y, "A Heart Curve")
```
